```python
import pandas as pd 
part1 = pd.read_csv("part1.csv")
part1.head(20)
```

```python
part2 = pd.read_csv("part2.csv")
part3 = pd.read_csv("part3.csv")
part4 = pd.read_csv("part4.csv")
part5 = pd.read_csv("part5.csv")
```

```python
sensor = pd.concat([part1, part2, part3, part4, part5], ignore_index=True)
```

```python
sensor
```

```python
# A quick look at factory_id column 
sensor['factory_id'] = sensor['factory_id'].astype('category')
sensor ['factory_id'].describe()
```

```python
top_categories = sensor['factory_id'].value_counts().head(10)
top_categories 
```

```python
# Filter categories that appear at least N times
min_count = 120
filtered_categories = sensor['factory_id'].value_counts()[sensor['factory_id'].value_counts() >= min_count]
print(filtered_categories)
```

```python
filtered_list = list(filtered_categories.index)
filtered_list
```

```python
filtered_sensor = sensor[sensor["factory_id"].isin(filtered_list)]
filtered_sensor
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

# For most counts 
grouped_avg = filtered_sensor.groupby("factory_id", observed = True)["sensor_value"].mean()

# Plotting
grouped_avg.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average Sensor Value by Factory")
plt.xlabel("Factory")
plt.ylabel("Average Sensor Value")
plt.xticks(rotation=0)
plt.show()
```

```python
#Voltage, most counts
filtered_categories = sensor['factory_id'].value_counts()[sensor['factory_id'].value_counts() == 120]
filtered_list = list(filtered_categories.index)
filtered_sensor = sensor[sensor["factory_id"].isin(filtered_list)]

grouped_avg = filtered_sensor.groupby("factory_id", observed = True)["voltage"].mean()

# Plotting
grouped_avg.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average Voltage by Factory")
plt.xlabel("Factory")
plt.ylabel("Average Voltage")
plt.xticks(rotation=45)
plt.show()
```

```python
# voltage, least counts
filtered_categories = sensor['factory_id'].value_counts()[sensor['factory_id'].value_counts() == 3]
filtered_list = list(filtered_categories.index)
filtered_sensor = sensor[sensor["factory_id"].isin(filtered_list)]

grouped_avg = filtered_sensor.groupby("factory_id", observed = True)["voltage"].mean()

# Plotting
grouped_avg.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average Voltage by Factory")
plt.xlabel("Factory")
plt.ylabel("Average Voltage")
plt.xticks(rotation=45)
plt.show()
```

```python
# wake-up counts, most counts
filtered_categories = sensor['factory_id'].value_counts()[sensor['factory_id'].value_counts() == 120]
filtered_list = list(filtered_categories.index)
filtered_sensor = sensor[sensor["factory_id"].isin(filtered_list)]

grouped_avg = filtered_sensor.groupby("factory_id", observed = True)["wakeup_counter"].mean()
grouped_avg = grouped_avg / 1000

# Plotting
grouped_avg.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average Wakeup Count by Factory")
plt.xlabel("Factory")
plt.ylabel("Average Wakeup Count(1000)")
plt.xticks(rotation=45)
plt.show()
```
